```python
from zipfile import ZipFile
import pandas as pd

with ZipFile('Reviews.csv.zip', 'r') as zip_ref:
    zip_ref.extractall()

df = pd.read_csv('Reviews.csv')

!pip install spacy
!python -m spacy download en_core_web_sm
!pip install nltk

import spacy
import nltk
from nltk.corpus import stopwords
import string

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

reviews = df['Text']
reviews = reviews.dropna()
reviews = reviews[:10000]

reviews_clean = reviews.str.lower().str.replace(f'[{string.punctuation}]', '', regex=True)

nlp = spacy.load("en_core_web_sm")

def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.text for token in doc if token.is_alpha and token.text not in stop_words]
    return tokens

def pos_tagging(text):
    doc = nlp(text)
    return [(token.text, token.pos_) for token in doc if token.is_alpha and token.text not in stop_words]

def named_entities(text):
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]

for i in range(100):
    print(f"Review {i+1}:")
    print("Tokens & POS:", pos_tagging(reviews_clean.iloc[i]))
    print("Entities:", named_entities(reviews_clean.iloc[i]))
    print()
```
